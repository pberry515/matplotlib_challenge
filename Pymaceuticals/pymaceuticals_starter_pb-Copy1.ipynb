## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset

combined_data_df = mouse_metadata.merge(study_results, how='left', on='Mouse ID', sort = False)

# Display the data table for preview

combined_data_df
```

```python
# Checking the number of mice.

total_mice = combined_data_df['Mouse ID'].count()
total_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate_mice = combined_data_df.loc[combined_data_df.duplicated(subset=['Mouse ID', 'Timepoint'])]
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. # difference between original and unique is 249, next
# step should create a df of 249

all_duplicate_mice = combined_data_df[combined_data_df.duplicated(['Mouse ID'])]
all_duplicate_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

clean_data_df=combined_data_df[combined_data_df['Mouse ID'].isin(duplicate_mice)==False]
clean_data_df
```

```python
# Checking the number of mice in the clean DataFrame.

cleaned_mice = clean_data_df['Mouse ID'].count()
cleaned_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each 
# regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

#Creating the variables

mean = clean_data_df.groupby('Drug Regimen')["Tumor Volume (mm3)"].mean()
median = clean_data_df.groupby('Drug Regimen')["Tumor Volume (mm3)"].median()
variance = clean_data_df.groupby('Drug Regimen')["Tumor Volume (mm3)"].var()
standard_deviation = clean_data_df.groupby('Drug Regimen')["Tumor Volume (mm3)"].std()
sem = clean_data_df.groupby('Drug Regimen')["Tumor Volume (mm3)"].sem()

#Creating the dataframe
summary_table_df = pd.DataFrame({'Mean': mean,
                                'Median':median,
                                'Variance': variance,
                                'Std Dev': standard_deviation,
                                'SEM': sem})
summary_table_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

single_group_by = clean_data_df.groupby('Drug Regimen')

summary_stats = single_group_by.agg(['mean','median','var','std','sem'])["Tumor Volume (mm3)"]

summary_stats

```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen


#Create the X and Y values
x_values=capo_grouped_plot['Weight (g)']
y_values=capo_grouped_plot['Tumor Volume (mm3)']

#Create the linear regression
(slope,intercept,rvalue,pvalue,stderr)=st.linregress(x_values, y_values)

#finding regression by x values times the slope and intercept
regress_value=x_values*slope + intercept

#Create the line equation

line_eq="y=" + str(round(slope,2)) + "x+" + str(round(intercept,2))

#Plot the scatter
plt.scatter(x_values,y_values)

# Plot the line
plt.plot(x_values,regress_value,"r-")

#Create Color for line
plt.annotate(line_eq,(6,10),fontsize=10,color='red')

#Create title and labels
plt.xlabel("Weight")
plt.ylabel("Tumor Volume")
plt.title("Weight vs. Avg Tumor Volume")

plt.savefig('linear_regression_chart.png')
```
